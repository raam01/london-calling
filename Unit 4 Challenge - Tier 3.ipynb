**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
properties_formatted.loc[:,["Borough","AreaCode"]]
```

```python
#Removing empty areas
properties_formatted=properties_formatted.dropna(subset=["AreaCode"])
properties_formatted.isna().sum()
```

```python
#boroughs currently in data
nb=list(properties_formatted["Borough"])
print(nb)
```

```python
#list of boroughs not needed
not_borough=["City of London",'Inner London','Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England']
```

```python
#removing data that are not a part of the 32 boroughs
properties_borough=properties_formatted[~properties_formatted.Borough.isin(not_borough)]
print(list(properties_borough["Borough"]))
print(len(properties_borough["Borough"]))
```

```python
#saving the list of boroughs
list_of_boroughs=list(properties_borough["Borough"])
print(list_of_boroughs)
```

```python
#melting the dataframe to required format for easy manipulation
boroughs=pd.melt(properties_borough,id_vars=["Borough","AreaCode"],var_name="date", value_name='avg price',)
boroughs
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#Visualizing the trend to see the avg price of one borough
Westminster=boroughs[boroughs["Borough"]=="Westminster"]
Westminster.plot(x="date",y="avg price")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Average price of Westminister over the years")
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

2. extract the year from each individual value in that column

```python
#Resample by year to get annual average price
tbased=boroughs.copy()
tbased1=tbased.groupby('Borough').apply(lambda x: x.set_index('date').resample('A').mean()).reset_index()
print(tbased1.head())
print(tbased1.shape)
```

```python
#Create a new year column for getting the required data easily
tbased1["year"]=tbased1["date"].dt.year
tbased1.head()
```

```python
#Plot the annual average price change across all boroughs 
import seaborn as sns
%matplotlib qt
sns.set(rc={'figure.figsize':(14.6,10.3)})
#f = plt.figure()
sns.lineplot(data=tbased1,x="year",y="avg price", hue="Borough",style="Borough",markers=True)
plt.legend(bbox_to_anchor=(1.005, 1), loc=2, borderaxespad=0.)
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.title("Line plot of average price by year for all boroughs")
plt.show()
#f.savefig("foo.pdf", bbox_inches='tight')
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Function that provides the ratio, average annual price in 1998 and 2018
def create_price_ratio(bor):
    price_1998=int(tbased1.loc[(tbased1["Borough"]==bor)&(tbased1["year"]==1998)]["avg price"].values[0])
    price_2018=int(tbased1.loc[(tbased1["Borough"]==bor)&(tbased1["year"]==2018)]["avg price"].values[0])
    ratio=round(price_1998/price_2018,3)
    return ratio,price_1998,price_2018
    #return ratio
```

```python
#Calling the function to get data of ratios and prices of those years
dict_price_ratios={}

for x in list_of_boroughs:
     dict_price_ratios[x]=create_price_ratio(x)   
print(dict_price_ratios)
```

```python
#Ordered dictionary 
dict_price_ratios=dict(sorted(dict_price_ratios.items(), key=lambda item: item[1]))
dict_price_ratios
```

```python
#Getting the list of top 3 and bottom 3 in the highest ratios data
top3=list(dict_price_ratios)[:3]
bottom3=list(dict_price_ratios)[-3:]
print(top3)
print(bottom3)
```

```python
#Exploring the 3 boroughs with the highest ratios
%matplotlib inline
top3_data=tbased1[tbased1["Borough"].isin(top3)]
sns.lineplot(data=top3_data,x="year",y="avg price", hue="Borough",style="Borough",markers=True)
plt.legend(bbox_to_anchor=(1.005, 1), loc=2, borderaxespad=0.)
plt.title("Avg price of 3 Boroughs with highest ratios")
plt.xlabel("Year")
plt.ylabel("Annual Average Price")
plt.show()
```

```python
#Exploring the 3 boroughs with the lowest ratios
%matplotlib inline
bottom3_data=tbased1[tbased1["Borough"].isin(bottom3)]
sns.lineplot(data=bottom3_data,x="year",y="avg price", hue="Borough",style="Borough",markers=True)
plt.legend(bbox_to_anchor=(1.005, 1), loc=2, borderaxespad=0.)
plt.title("Avg price of 3 Boroughs with lowest ratios")
plt.xlabel("Year")
plt.ylabel("Annual Average Price")
plt.show()
```
